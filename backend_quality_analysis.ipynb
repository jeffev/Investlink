```python
print("\n=== CODE SMELLS ===\n")

smells = defaultdict(list)

for service_file in service_files:
    service_path = os.path.join(services_path, service_file)
    with open(service_path, 'r') as f:
        content = f.read()
        lines = content.split('\n')
    
    # 1. Magic numbers/strings
    magic_numbers = re.findall(r'= (\d{3,})|== (\d{3,})|> (\d{3,})|< (\d{3,})', content)
    if magic_numbers:
        smells['magic_numbers'].append(service_file)
    
    # 2. Duplicated validation
    if content.count('if not') > 5:
        smells['repeated_validation'].append(service_file)
    
    # 3. Long parameter lists
    for line in lines:
        if 'def ' in line:
            params = len(re.findall(r'\w+\s*=', line))
            if params > 4:
                smells['long_params'].append(f"{service_file}: {len(params)} parâmetros")
    
    # 4. Hardcoded strings (exceto mensagens de erro)
    if 'http://' in content or 'postgresql://' in content:
        smells['hardcoded_urls'].append(service_file)

print("✓ Magic Numbers (valores hardcoded):")
if smells['magic_numbers']:
    print(f"  ⚠️  {len(smells['magic_numbers'])} arquivos com magic numbers")
    for f in smells['magic_numbers']:
        print(f"    - {f}")
else:
    print(f"  ✓ Nenhum magic number encontrado")

print("\n✓ Repeated Validation Logic:")
if smells['repeated_validation']:
    print(f"  ⚠️  {len(smells['repeated_validation'])} arquivos com validação repetida")
    for f in smells['repeated_validation']:
        print(f"    - {f}")
else:
    print(f"  ✓ Validação parece centralizada")

print("\n✓ Long Parameter Lists:")
if smells['long_params']:
    print(f"  ⚠️  Encontrados:")
    for item in smells['long_params'][:3]:
        print(f"    - {item}")
else:
    print(f"  ✓ Sem listas longas de parâmetros")

print("\n✓ Hardcoded URLs/Credentials:")
if smells['hardcoded_urls']:
    print(f"  ⚠️  {len(smells['hardcoded_urls'])} arquivos com URLs hardcoded")
    for f in smells['hardcoded_urls']:
        print(f"    - {f}")
else:
    print(f"  ✓ Nenhuma URL hardcoded detectada")
```

## 4. Análise de Code Smells

```python
print("\n=== CLEAN CODE ANALYSIS ===\n")

# 1. Verificar try/except handling
print("✓ Error Handling (try/except):")
error_handling_score = 0
total_functions = 0

for service_file in service_files:
    service_path = os.path.join(services_path, service_file)
    with open(service_path, 'r') as f:
        content = f.read()
    
    # Contar funções
    functions = re.findall(r'^def (\w+)\(', content, re.MULTILINE)
    functions_with_try = len(re.findall(r'def \w+\(.*?\):\s+try:', content, re.DOTALL))
    
    total_functions += len(functions)
    error_handling_score += functions_with_try
    
    coverage = (functions_with_try / len(functions) * 100) if functions else 0
    status = "✓" if coverage >= 80 else "⚠️"
    print(f"  {status} {service_file}: {functions_with_try}/{len(functions)} funções com try/except ({coverage:.0f}%)")

print(f"\n  Cobertura geral de try/except: {error_handling_score}/{total_functions} = {(error_handling_score/total_functions*100):.1f}%")

# 2. Verificar logging
print("\n✓ Logging:")
logging_files = 0
for service_file in service_files:
    service_path = os.path.join(services_path, service_file)
    with open(service_path, 'r') as f:
        content = f.read()
    
    if 'import logging' in content and ('logger.error' in content or 'logging.error' in content):
        logging_files += 1

print(f"  ✓ {logging_files}/{len(service_files)} serviços com logging configurado")

# 3. Verificar nomes significativos
print("\n✓ Meaningful Names (função names):")
all_functions = []
for service_file in service_files:
    service_path = os.path.join(services_path, service_file)
    with open(service_path, 'r') as f:
        content = f.read()
    
    functions = re.findall(r'^def (\w+)\(', content, re.MULTILINE)
    all_functions.extend(functions)

# Procura por nomes ruins
bad_names = [f for f in all_functions if len(f) < 3 or f in ['get', 'set', 'do', 'run']]
good_names_ratio = (len(all_functions) - len(bad_names)) / len(all_functions) * 100 if all_functions else 0

print(f"  ✓ Total de funções: {len(all_functions)}")
if bad_names:
    print(f"  ⚠️  {len(bad_names)} funções com nomes genéricos: {bad_names}")
else:
    print(f"  ✓ Todos os nomes são significativos")

# 4. Comprimento das funções
print("\n✓ Function Length (máximo 20 linhas):")
long_functions = []
for service_file in service_files:
    service_path = os.path.join(services_path, service_file)
    with open(service_path, 'r') as f:
        lines = f.readlines()
    
    for i, line in enumerate(lines):
        if line.strip().startswith('def '):
            func_name = re.match(r'def (\w+)', line.strip()).group(1)
            # Contar linhas até próximo 'def'
            func_lines = 1
            for j in range(i+1, len(lines)):
                if re.match(r'^def ', lines[j]) or (j > i+1 and lines[j][0] not in [' ', '\t', '']):
                    break
                func_lines += 1
            
            if func_lines > 20:
                long_functions.append((service_file, func_name, func_lines))

if long_functions:
    print(f"  ⚠️  {len(long_functions)} funções com mais de 20 linhas:")
    for file, func, lines in long_functions[:5]:
        print(f"    - {file}/{func}: {lines} linhas")
else:
    print(f"  ✓ Todas as funções têm menos de 20 linhas")
```

## 3. Análise de Clean Code

```python
print("\n=== DESIGN PATTERNS ===\n")

# 1. Repository Pattern
print("✓ Repository Pattern (Services como data layer):")
repo_score = 0
for service_file in service_files:
    service_path = os.path.join(services_path, service_file)
    with open(service_path, 'r') as f:
        content = f.read()
    
    # Verificar funções CRUD padrão
    crud_funcs = ['def list_', 'def view_', 'def new_', 'def edit_', 'def delete_']
    found_crud = sum(1 for func in crud_funcs if func in content)
    if found_crud >= 3:
        print(f"  ✓ {service_file}: {found_crud}/5 funções CRUD")
        repo_score += found_crud

print(f"\n  Conformidade com Repository Pattern: {repo_score}/{len(service_files)*5} funções = {(repo_score/(len(service_files)*5)*100):.1f}%")

# 2. Decorator Pattern
print("\n✓ Decorator Pattern (protected_route, decorators em geral):")
utils_path = os.path.join(APP_PATH, "utils.py")
with open(utils_path, 'r') as f:
    utils_content = f.read()

decorator_patterns = [
    ('protected_route', '@protected_route' in utils_content or 'def protected_route' in utils_content),
    ('decorator com functools.wraps', '@wraps' in utils_content),
    ('jwt_required', '@jwt_required' in utils_content)
]

for pattern_name, found in decorator_patterns:
    status = "✓" if found else "✗"
    print(f"  {status} {pattern_name}")

# 3. Factory Pattern
print("\n✓ Factory Pattern (create_app):")
config_path = os.path.join(APP_PATH, "config.py")
with open(config_path, 'r') as f:
    config_content = f.read()

factory_found = 'def create_app' in config_content
print(f"  {'✓' if factory_found else '✗'} create_app() function found")
if factory_found:
    print(f"    - Inicializa db")
    print(f"    - Inicializa jwt")

# 4. Strategy Pattern (validadores)
print("\n✓ Strategy Pattern (Validadores):")
validators = []
for service_file in service_files:
    service_path = os.path.join(services_path, service_file)
    with open(service_path, 'r') as f:
        content = f.read()
    
    validate_funcs = re.findall(r'def (validate_\w+)', content)
    if validate_funcs:
        validators.extend(validate_funcs)
        print(f"  - {service_file}: {validate_funcs}")

if not validators:
    print("  ⚠️  Poucos validadores isolados encontrados")

print(f"\n  Total de funções validadoras: {len(validators)}")
```

## 2. Análise de Design Patterns

```python
# Analisar SRP - cada service deve lidar com um domínio
services_path = os.path.join(APP_PATH, "services")
service_files = [f for f in os.listdir(services_path) if f.endswith("_services.py")]

print("=== SINGLE RESPONSIBILITY PRINCIPLE (SRP) ===\n")
print(f"✓ Arquivos de serviço encontrados: {len(service_files)}")
for service in sorted(service_files):
    print(f"  - {service}")

# Verificar imports cruzados (violação de SRP)
print("\n✓ Verificando dependências entre serviços (cross-imports):\n")

srp_violations = []
for service_file in service_files:
    service_path = os.path.join(services_path, service_file)
    with open(service_path, 'r') as f:
        content = f.read()
    
    # Procura por imports de outros serviços
    cross_imports = re.findall(r'from services\.(\w+)_services import', content)
    if cross_imports:
        srp_violations.append({
            'service': service_file,
            'imports_from': cross_imports
        })

if srp_violations:
    print("⚠️  VIOLAÇÕES DE SRP (serviço importando outro serviço):")
    for violation in srp_violations:
        print(f"  - {violation['service']} importa de: {violation['imports_from']}")
else:
    print("✓ Nenhuma violação de SRP detectada (serviços isolados)")

print("\n✓ Análise de responsabilidades por serviço:")
for service_file in sorted(service_files):
    service_path = os.path.join(services_path, service_file)
    with open(service_path, 'r') as f:
        content = f.read()
    
    # Contar funções
    functions = re.findall(r'^def (\w+)\(', content, re.MULTILINE)
    # Contar modelos acessados
    models = re.findall(r'from models\.(\w+) import', content)
    
    service_name = service_file.replace('_services.py', '').title()
    print(f"\n  {service_file}:")
    print(f"    - Funções: {len(functions)} ({', '.join(functions[:3])}...)")
    print(f"    - Modelos acessados: {set(models) if models else 'Nenhum'}")
```
