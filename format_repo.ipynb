```python
# Notebook: Aplicar formatação automática
# Section 1: Instalar dependências de formatação (exemplo)
# Section 5: Executar formatadores via terminal e células

import subprocess
import sys

commands = [
    "python -m pip install --user black",
    "python -m black .",
    "git config user.email 'devnull@example.com'",
    "git config user.name 'Investlink Bot'",
    "git add -A",
    "git commit -m 'style: format code with black' || echo 'No changes to commit'",
]

for cmd in commands:
    print('\n>>> Running:', cmd)
    proc = subprocess.run(cmd, shell=True, check=False, stdout=subprocess.PIPE, stderr=subprocess.STDOUT)
    out = proc.stdout.decode('utf-8', errors='ignore')
    print(out)
    print('Return code:', proc.returncode)
```
